```python
import pandas as pd
data = [12, 24, 36, 48, 60]
index = ['A', 'B', 'C', 'D', 'E']
s = pd.Series(data, index=index)
element_loc = s.loc['C']
print("Элемент с индексом 'C':", element_loc)
element_iloc = s.iloc[2]
print("Третий элемент:", element_iloc)
```

```python
import pandas as pd
import numpy as np
data = np.array([4, 9, 16, 25, 36, 49, 64])
s = pd.Series(data)
result = s[s > 20]
print("Исходный Series:")
print(s)
print("\nЭлементы > 20:")
print(result)
```

```python
import pandas as pd
import numpy as np
data = [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]
s = pd.Series(data)
has_nans = s.isna().any()
print("Есть ли пропущенные значения:", has_nans)
nan_indices = s.index[s.isna()]
print("\nИндексы пропущенных значений:", nan_indices.tolist())
```

```python
import pandas as pd
import numpy as np
data = [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]
s = pd.Series(data)
mean_value = s.mean()
s_filled = s.fillna(mean_value)
print("Исходный Series:")
print(s)
print("\nСреднее значение непустых элементов:", mean_value)
print("\nSeries после замены NaN:")
print(s_filled)
```

```python
import pandas as pd
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])
result = s1 + s2
print("Результат сложения без обработки:")
print(result)
result_filled = result.fillna(0)
print("\nРезультат после замены NaN на 0:")
print(result_filled)
```

```python
import pandas as pd
import numpy as np
numbers = [2, 4, 6, 8, 10]
s = pd.Series(numbers)
s_sqrt = s.apply(np.sqrt)
print("Исходный Series:")
print(s)
print("\nSeries с квадратными корнями:")
print(s_sqrt)
```

```python
import pandas as pd
import numpy as np 
s = pd.Series(np.random.randint(50, 151, size=20))
stats = {
    'Сумма': s.sum(),
    'Среднее': s.mean(),
    'Минимум': s.min(),
    'Максимум': s.max(),
    'Стандартное отклонение': s.std()
}
print("Сгенерированный Series:")
print(s)
print("\nСтатистические показатели:")
for key, value in stats.items():
    print(f"{key}: {value:.2f}" if isinstance(value, float) else f"{key}: {value}")
```

```python
import pandas as pd
s = pd.Series([10, 20, 30, 40, 50, 60], 
              index=['A', 'B', 'A', 'C', 'D', 'B'])
is_unique = s.index.is_unique
print(f"Индексы уникальные: {is_unique}")
if not is_unique:
    s_grouped = s.groupby(level=0).sum()
    print("\nРезультат после группировки:")
    print(s_grouped)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dates = pd.date_range(start='2024-03-01', end='2024-03-30', freq='D')
values = np.random.randint(50, 151, size=len(dates))
ts = pd.Series(values, index=dates, name='Случайные значения')
plt.figure(figsize=(12, 6))
ts.plot(color='blue', linewidth=2, marker='o', markersize=5)
plt.title('Случайные значения с 1 по 30 марта 2024 года', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Значение', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Индивидуальное задание №5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dates = pd.date_range(start='2024-03-10', end='2024-03-20')
temperatures = np.random.uniform(low=-5, high=10, size=len(dates))
temperature_series = pd.Series(temperatures, index=dates, name='Температура')
selected_data = temperature_series['2024-03-15':'2024-03-17']
selected_data.index = pd.DatetimeIndex(selected_data.index)
plt.figure(figsize=(10, 5))
selected_data.plot(marker='o', linestyle='-')
plt.title('Температура 15-17 марта 2024 года')
plt.ylabel('Температура (°C)')
plt.xlabel('Дата')
plt.grid(True)
plt.show()
```
